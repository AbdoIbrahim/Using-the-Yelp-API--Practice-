```python
import pandas as pd 
import json
with open('/Users/abibrahim/.secret/yelp_api.json') as f:
    login = json.load(f)
login.keys()
```

```python
# import the YelpAPI Class
from yelpapi import YelpAPI
# Create an instance with your key
yelp_api = YelpAPI(login['api-key'], timeout_s=5.0)
yelp_api
```

```python
# use our yelp_api variable's search_query method to perform our API call
shawarma_results = yelp_api.search_query(location='NY, NY',
                                       term='Shawarma')
print(type(shawarma_results))
shawarma_results.keys()
```

```python
shawarma_0 = yelp_api.search_query(location='NY, NY',
                                       term='Shawarma')
type(shawarma_results)
shawarma_results.keys()
```

```python
shawarma0 = pd.DataFrame(shawarma_0['businesses'])
shawarma0.head(2)
```

```python
shawarma_20 = yelp_api.search_query(location='NY, NY',
                                       term='Shawarma',
                               offset = 20)
```
